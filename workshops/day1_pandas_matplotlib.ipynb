```python
# Add and remove rows and columns

# Add: pd.concat
# Remove: df.drop
```

```python
df1 = df.loc[(df['Timestamp'] < '2018-06-03') &
       (df['Timestamp'] >= '2018-06-02')]

df2 = df.loc[(df['Timestamp'] < '2018-06-10') &
       (df['Timestamp'] >= '2018-06-09')]

#print(df1.info())
#print(df2.info())

df3 = pd.concat([df1, df2], axis=0)
print(df3.info())
```

```python
# Remove rows
# df.drop(indices, axis=0)

# drop the June 9th entries from df3
# Or: indices = df2.index
indices = df3.loc[(df['Timestamp'] >= '2018-06-09') & 
        (df['Timestamp'] < '2018-06-10')].index

df4 = df3.drop(indices, axis=0)
df4.info()
```

```python
# Add columns
# pd.concat([columns], axis=1)

df2 = df.loc[(df['Timestamp'] < '2018-06-03') &
       (df['Timestamp'] >= '2018-06-02'), # row_selector
       ['Low', 'High']] # col_selector
print(df2.head())

# Let's say we want to put the +1 week values side by side
df1 = df.loc[(df['Timestamp'] < '2018-06-10') &
       (df['Timestamp'] >= '2018-06-09'),
       ['Low', 'High']]

print(df1.head())
```

```python
df3 = df1.reset_index() # reset index so we can align the rows
df4 = df2.reset_index()
print(df3.head())
print(df4.head())
```

```python
# when adding columns, the index must be aligned

# df.loc[:, col_selector] # select all rows, some columns

df5 = pd.concat([
   df3.loc[:, ['Low', 'High']], # first dataframe
   df4.loc[:, ['Low', 'High']] # second dataframe
], axis=1)

df5.columns = ['Low', 'High', 'Low+1wk', 'High+1wk']

df5.head()
```

```python
# Remove columns
# df.drop([columns], axis=1)

df6 = df5.drop(['Low', 'Low+1wk'], axis=1)
df6
```

```python
# plot Timestamp vs. Open price
# 15, 10 or 10, 8 are reasonable sizes
fig, ax = plt.subplots(figsize=(10, 8)) # width, height

ax.plot(df['Timestamp'], df['Open'])
fig.autofmt_xdate() # make dates look nicer

ax.set_title('Open price')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# or set all in 1 call
#ax.set(title='Open price', xlabel='Date', ylabel='Price')
```

```python
# 2 overlay plots

# plot Timestamp vs. High price and Low  price
fig, ax = plt.subplots(figsize=(10, 8)) # width, height

df1 = df.loc[df['Timestamp'] < '2018-06-02'] # zoom in

ax.plot(df1['Timestamp'], df1['Low'],
        color='orange',
        label='Low') # reuse canvas

ax.plot(df1['Timestamp'], df1['High'],
        color='purple', alpha=0.5,
        label='High') # transparency

ax.legend()

fig.autofmt_xdate() # make dates look nicer

ax.set_title('High and low prices')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
```

```python
# 2 side-by-side plots

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

ax[0].plot(df['Timestamp'], df['Open'])
ax[0].set(title='Open')

ax[1].plot(df['Timestamp'], df['Volume_(BTC)'])
ax[1].set(title='Volume (Bitcoin currency)')

fig.autofmt_xdate()
```

```python
# 4 side-by-side plots

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
ax = ax.flatten()

ax[0].plot(df['Timestamp'], df['Open'])
ax[0].set(title='Open')

ax[1].plot(df['Timestamp'], df['Volume_(BTC)'])
ax[1].set(title='Volume (Bitcoin currency)')

ax[2].plot(df['Timestamp'], df['Close'])
ax[2].set(title='Close')

fig.autofmt_xdate()
```
