## Pandas

```python
# 1. Data Engineering
#    What: load, transform / process, visualise data
#    Use: Pandas, Matplotlib (NumPy)

# load data
# 1. fix the header
ecg = pd.read_csv('../data/ecg/echocardiogram.data',
                  error_bad_lines=False,
                  names=['survival', 'still_alive',
                         'age_at_heart_attack', 'pericardial_effusion',
                         'fractional_shortening', 'epss',
                         'lvdd', 'wall_motion_score',
                         'wall_motion_index', 'mult', 'name',
                         'group', 'alive_at_1'],
                  header=None, # header=None to keep first row
                  na_values=['?']) # add ? to be NaN value

#print(ecg.info()) # before change

# minimal process (removing ? values)
# 2. remove ? leads to half gone
#ecg.dropna(inplace=True) # drop NaN from dataframe, inplace
#print(ecg.info()) # after removing ?

ecg_fixed = ecg.interpolate()
#print(ecg_fixed.info()) # after change

# visualize
ecg2 = ecg_fixed.loc[:, (ecg_fixed.columns != 'name') &
                        (ecg_fixed.columns != 'group')]
ecg2.info()
```

```python
# Scatterplot Matrix
from pandas.plotting import scatter_matrix

scatter_matrix(ecg2, figsize=(30, 30), alpha=0.9)
plt.show()
```
